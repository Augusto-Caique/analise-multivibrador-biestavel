### 1. Cálculo da Malha da Base (Q1 Saturado)

Quando o circuito é alimentado ($6\text{V}_\text{CC}$) uma pequena corrente passa para as bases de ambos transistores por um pequeno momento, mas o transistor que possui maior ganho (Vamos considerar $Q1$) vai conduzir uma a corrente maior em seu coletor em relaçao ao segundo transistor ($Q2$) e dessa forma a corrente da base do segundo transistor é "cortada".

Analisando a malha da base de $Q1$:

$$V_{CC} - V_{D2} - I_{B1} \cdot (R_2 + R_4) - V_{BE1} = 0  $$


Isolando a corrente de base ($I_{B1}$):

$$ I_{B1} = \frac {V_{CC} - V_{D2} - V_{BE1}}{R_2 + R_4}$$


Desenvolvendo o cáculo:

$$ I_{B1} = \frac {6 \text{V} - 2 \text{V} - 0.7 \text{V}} {1 \text{k} \Omega + 10 \text{k} \Omega}$$

$$  $$

$$ I_{B1} = \frac {3.3 \text{V}} {11 \text{k} \Omega} = 0.3 \text{mA}$$
